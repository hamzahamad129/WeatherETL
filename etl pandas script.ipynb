```python
import requests
```

```python
# Extraction of data
url_path='https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&hourly=temperature_2m'
def get_weather_data(url_path):
    response = requests.get(url_path)
    if response.status_code == 200:
        return response.json()
    else:
        return {"Error : Failed to retrieve data"}
    
data=get_weather_data(url_path)
```

```python
import pandas as pd
hourly_data = data["hourly"]
df= pd.DataFrame(hourly_data)
df.head(100)
```
